## Imports

import random

## Functions
### `genRandom`
```python
def genRandom():
    return random.uniform(-10, 10)
```
- **`genRandom`**: Generates a random floating-point number between -10 and 10.

### `summation`
```python
def summation(x, w, offset):
    result = 0
    for i in range(0, 2):
        result = result + x[i + offset] * w[i]
        print("Values of summation --- ", " Input: ", x[i + offset], " Weight: ", w[i])
    if(condition):
        print("summation result", result)
    return result
```
- **`summation`**: Calculates a weighted sum of elements in `x` based on the weights in `w` and prints the intermediate values.
  - **`x`**: List of input values.
  - **`w`**: List of weights.
  - **`offset`**: Index offset for selecting elements from `x`.

## Variables and Initialization
```python
offset = [0, 2, 4]
x = [0, 0, 0, 1, 1, 1]
w = [0, 0]
s = 0
y = 0

```
- **`offset`**: List of index offsets used in the `summation` function.
- **`x`**: List of input values.
- **`w`**: List of weights initialized to zero.
- **`s`**: Variable to store the summation result.
- **`y`**: Variable to store the activation output (0 or 1).
- **`results`**: List to track the results of different cases.
- **`condition`**: Boolean flag to control the loop execution.
