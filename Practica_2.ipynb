````python
import json
import random
from art import *

##Auxiliar functions
def generatePrice():
    result = (random.random()*random.randint(20,500))
    while (result < 0.1):
         result == (random.random()*random.randint(20,500))

    return result

def generateQuantity():
     return random.randint(1,100)

##Classes
class manageProducts:
    def __init__ (self, ProductsLT):
        self.ProductsLT = [()]
        self.ProductsLT = ProductsLT

    def __str__(self):
         jsonString = json.dumps(self.__dict__, ensure_ascii=False)
         return (
              f"{jsonString}"
         )

    ##CRUD functions
    def listProducts (self):
        print("-"*51)
        print(f"{'ID':<3}{'Product':<30}{'Price $ ':10}{'Quantity':10}")
        print("-"*51)
        for index, product in enumerate(self.ProductsLT):
             print(
                 f"{index:<3}{product[0]:<30}"
                 f"{product[1]:<10}"
                 f"{product[2]:<10}"
                )
        print("-"*51)


    def addProduct(self,name,price,quantity):
        self.ProductsLT.append((name, price, quantity))
        self.listProducts()

    def updateQuantity(self, id):
        print("\nBuy - 1, Sell - 2")
        selectedOption = int(input("Select Option: "))

        if(selectedOption == 1 and self.ProductsLT[id][2] < 5):
            quantityBuy = int(input("Quantity to buy: "))
            self.ProductsLT.insert(id, (self.ProductsLT[id][0],self.ProductsLT[id][1], self.ProductsLT[id][2] + quantityBuy))
            self.ProductsLT.pop(id + 1)
            self.listProducts()

        if(selectedOption == 1 and self.ProductsLT[id][2] > 5):
            print("\nThere is enough of this product")

        if(selectedOption == 2 and self.ProductsLT[id][2] > 0):
            quantitySell = int(input("Quantity to sell: "))
            if(quantitySell < self.ProductsLT[id][2]):
                self.ProductsLT.insert(id, (self.ProductsLT[id][0],self.ProductsLT[id][1], self.ProductsLT[id][2] - quantitySell))
                self.ProductsLT.pop(id + 1)
                self.listProducts()
            else:
                print("\nThere is not enough of this product")

        if(selectedOption == 2 and self.ProductsLT[id][2] < 0):
            print("\nThere is not enough of this product")

    def deleteProduct(self, id):
         self.ProductsLT.pop(id)
         self.listProducts()

##Methods
def main():
    status = True
    selection = ""
    tprint("             DOOM\nWORKSHOP", font="doom")
    productos = manageProducts([("Brake Pads",round(generatePrice(),2),generateQuantity()),
                                ("Alternator",round(generatePrice(),2),generateQuantity()),
                                 ("Oil Filter",round(generatePrice(),2),generateQuantity()),
                                 ("Spark Plugs",round(generatePrice(),2),generateQuantity()),
                                 ("Brake liquid",round(generatePrice(),2),generateQuantity())]
                                 )
    ##Program Loop
    while(status):
        print(
        f"\nSelect an option \n"
        f"1: Get stock \n2: Add a new product \n3: Modify product quantity (ID) \n4: Delete an element (Name)\n"
        f"0: Exit"
        )

        selection = input("\nSelect an option: ")
        print()

        ##Switch Case
        match selection:
            case "1":
                productos.listProducts()

            case "2":
                print(f"Insert product: ")
                productos.addProduct(input("Insert name: "), input("Insert price: "), input("Insert quantity: "))

            case "3":
                print(f"Modify product quantity: ")
                productos.updateQuantity(int(input("Insert ID: ")))

            case "4":
                  print(f"Delete an element by ID: ")
                  productos.deleteProduct(int(input("Insert ID: ")))

            case "0":
                status = False

##Call to main method
if __name__ == "__main__":
        main()
````

````python
import random
from faker import Faker

# Define fake
fake = Faker()

# Auxiliar functions
def generateEmployees(quantity, employees):
    for x in range(quantity):
        employees.append(dict(Name=fake.name(),
                              hoursWorked=random.randint(24, 80),
                              rate=(random.random() + random.randint(1, 3)) * random.randint(250, 600)))

# Classes
class manageProjects:
    def __init__(self, quantity):
        self.offset = [0, 20, 40]
        self.employees = []
        self.projects = [[None] * 21 for _ in range(3)]
        generateEmployees(quantity, self.employees)
        print(self.employees)
        self.projects[0][0] = fake.company()
        self.projects[1][0] = fake.company()
        self.projects[2][0] = fake.company()
        for x in range(3):
            for y in range(self.offset[x], 20 + self.offset[x]):
                self.projects[x][y - self.offset[x] + 1] = self.employees[y]

    # Functions:
    def listProjects(self):
        print("-" * 51)
        print(f"{'ID':<3}{'Project Name':<60}")
        print("-" * 51)
        for x in range(len(self.projects)):
            print(f"{x:<3}{self.projects[x][0]:<60}")
        print("-" * 51)

    def getProjectID(self):
        while True:
            try:
                projectID = int(input("\nInsert Project ID: "))
                if 0 <= projectID < len(self.projects):
                    return projectID
                else:
                    print(f"Invalid Project ID. Please enter a number between 0 and {len(self.projects)-1}.")
            except ValueError:
                print("Invalid input. Please enter a valid integer.")

    def listEmployeesByProject(self):
        self.listProjects()
        projectID = self.getProjectID()
        print("-" * 61)
        print(f"{'ID':<6}{'Name':<30}{'Hours Worked':<15}{'Rate':<15}")
        print("-" * 61)
        for x in range(1, 21):
            if x < len(self.projects[projectID]):
                employee = self.projects[projectID][x]
                emp_id = str(random.randint(1111, 9999))
                name = employee['Name']
                hours_worked = employee['hoursWorked']
                rate = employee['rate']
                print(f"{emp_id:<6}{name:<30}{hours_worked:<15}{rate:<15.2f}")
        print("-" * 61)

    def totalHoursByProject(self):
        self.listProjects()
        projectID = self.getProjectID()
        result = 0
        print("-" * 21)
        print(f"{'ID':<6}{'Total Hours':<15}")
        print("-" * 21)
        for x in range(1, 21):
            if x < len(self.projects[projectID]):
                employee = self.projects[projectID][x]
                result += employee['hoursWorked']
        print(f"{projectID:<6}{result:<15}")
        print("-" * 21)

    def totalCostByProject(self):
        self.listProjects()
        projectID = self.getProjectID()
        result = 0
        print("-" * 21)
        print(f"{'ID':<6}{'Total Cost':<15}")
        print("-" * 21)
        for x in range(1, 21):
            if x < len(self.projects[projectID]):
                employee = self.projects[projectID][x]
                result += employee['hoursWorked'] * employee['rate']
        print(f"{projectID:<6} $ {result:<15.2f}")
        print("-" * 21)

# Methods
def main():
    # Auxiliary variables
    status = True
    # Create employees
    projects = manageProjects(60)

    # Program Loop
    while status:
        print(
            f"\nSelect an option \n"
            f"1: List projects \n2: List employees by project (Project ID) \n3: Total hours by project (Project ID) \n4: Total cost by project (Project ID)\n"
            f"0: Exit"
        )

        selection = input("\nSelect an option: ")
        print()

        # Switch Case
        match selection:
            case "1":
                projects.listProjects()
            case "2":
                projects.listEmployeesByProject()
            case "3":
                projects.totalHoursByProject()
            case "4":
                projects.totalCostByProject()
            case "0":
                status = False

# Call to main method
if __name__ == "__main__":
    main()
````
