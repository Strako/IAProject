````python
from tabulate import tabulate

xA = [0,0,0,0]
xB = [0,0,0,0]
xC = [0,0,0,0]
yC = [0,0,0,0]

xSummation =[]
ySummation =[]
xStyled = []
yStyled = []
summationHeaders = ["X1", "W1", "X2", "W2", "Bias", "∑", "Y"]
resultHeaders = ["A","B","C"]
default = []
andBiasA = 0
andBiasB = -3
orBias = -0.5
w = [3, -2.5]
v =[1,1]
offset= [0,2,4,6]

for i in range(0,4):
  default.append([f"A{i +1}",f"B{i + 1}"])

print(tabulate(default, headers=["A","B"], tablefmt="pretty", numalign="center", stralign="center"))

for i in range(0, 4):
  xA[i] = int(input(f"Ingresa el valor A {i + 1} : "))
  xB[i] = int(input(f"Ingresa el valor B {i + 1} : "))

values = [xA[0],xB[0],xA[1],xB[1],xA[2],xB[2],xA[3],xB[3],]


#Linear activation function
def linear(input):
  if(input > 0 ):
    return 1
  else:
    return 0

#AND Gate Summation function
def summationAND(values ,w, offset):
  result = 0
  usedBias = 0
  if (0 + offset < len(values) -1) and (values[0 + offset] == 1 and values[1 + offset] == 0):
    usedBias = andBiasB
    result = usedBias
  else:
    usedBias=andBiasA
    result = usedBias

  for i in range(0,2):
    result = result + values[i + offset] * w[i]

  xSummation.append([values[0 + offset],w[0],values[1 + offset],w[1],usedBias, result, linear(result)])
  return linear(result)

#OR Gate Summation function
def summationOR(values ,v, offset):
  result = orBias
  for i in range(0,2):
    result = result + values[i + offset] * v[i]
  ySummation.append([values[0 + offset],v[0],values[1 + offset],v[1],orBias, result, linear(result)])
  return linear(result)

for x in range(0,4):
  xC[x] = summationAND(values,w,offset[x])
  xStyled.append([xA[x], xB[x], xC[x]])
  yC[x] = summationOR(values,v,offset[x])
  yStyled.append([xA[x], xB[x], yC[x]])


#Print table
print("\nAND Summation and linear activation function")
print(tabulate(xSummation, headers=summationHeaders, tablefmt="pretty", numalign="center", stralign="center"))

print("\nAND Logic Gate")
print(tabulate(xStyled, headers=resultHeaders, tablefmt="pretty", numalign="center", stralign="center"))

print("\nOR Summation and linear activation function")
print(tabulate(ySummation, headers=summationHeaders, tablefmt="pretty", numalign="center", stralign="center"))

print("\nOR Logic Gate")
print(tabulate(yStyled, headers=resultHeaders, tablefmt="pretty", numalign="center", stralign="center"))


````
